```python
import geopandas as gpd
from shapely.geometry.multipolygon import MultiPolygon
import pandas as pd
```

```python
# 2020年数据里的经纬度跟后面数据的经纬度有重复的么
path = '/Users/anora/Library/CloudStorage/Dropbox-TeamMG/Wanru Wu/Cloudseeding_Anora/operation/cleaned_data.csv'
df = pd.read_csv(path)

# split
df_2020 = df[df['date'].astype(str).str.contains('2020')]
df_later = df[~df['date'].astype(str).str.contains('2020')]

# unique coordinate sets
coords_2020 = set(zip(df_2020['lon'].round(2), df_2020['lat'].round(2)))
coords_later = set(zip(df_later['lon'].round(2), df_later['lat'].round(2)))
intersect = coords_2020.intersection(coords_later)
print(len(intersect))
print(len(coords_2020))
print(len(coords_later))
print(intersect)
```

```python
# 2020年 同一个镇子不同的经纬度比例有多少？
path = '/Users/anora/Library/CloudStorage/Dropbox-TeamMG/Wanru Wu/Cloudseeding_Anora/operation/cleaned_data.csv'
df = pd.read_csv(path)

# keep only 2020
df_2020 = df[df['date'].astype(str).str.contains('2020')]

# round to 2 decimals
df_2020['lon'] = df_2020['lon'].round(2)
df_2020['lat'] = df_2020['lat'].round(2)

# count unique lon/lat pairs in each town
counts = df_2020.groupby('location_o')[['lon','lat']].apply(lambda x: len(x.drop_duplicates()))
print(len(counts[counts>1]))
print(counts[counts>1])
print(len(counts[counts>1])/len(counts))
```
