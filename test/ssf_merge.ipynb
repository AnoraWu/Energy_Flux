```python
import pandas as pd

# 模拟原始数据
df1 = pd.DataFrame({
    "Cloud_Layers": [0,1,0,1],
    "col1": [1, 1, 3, 3],
    "col2": [2, 2, 4, 4],
    "col3": [3, 6, None, 12]
})
df1
```

```python
# 透视
df1 = df1.groupby(['Cloud_Layers','col2','col1']).mean()
df1 = df1.reset_index()

df1_wide = df1.pivot(
    index=["col1", "col2"], 
    columns="Cloud_Layers"
)

# 展平列名
df1_wide.columns = [f"{col}_layer{layer}" for col, layer in df1_wide.columns]

df1_wide = df1_wide.reset_index()
df1_wide
```

```python
# 展平列名
df1_wide.columns = [f"{col}_{layer}" for col, layer in df1_wide.columns]
df1_wide = df1_wide.reset_index(drop=True)
df1_wide
```
